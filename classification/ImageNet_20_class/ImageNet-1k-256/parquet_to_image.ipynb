```python
def single_pic_recover(row,target_folder,N,gray_flag,save_as):

    image_data = row['image']

    image_label = row['label']

    target_class_folder = os.path.join(target_folder,str(image_label))

    if not os.path.exists(target_class_folder):
        os.makedirs(target_class_folder)

    image_bytes = image_data['bytes']
    origin_name = image_data['path']

    image = Image.open(io.BytesIO(image_bytes))

    new_size = (N, N)
    image = image.resize(new_size)

    if gray_flag == 1:
        image = image.convert('L')

    if save_as == 'png':
        name,extension = os.path.splitext(origin_name)
        new_name = name + '.png'
        image.save(os.path.join(target_folder,str(image_label),new_name))
    else:
        image.save(os.path.join(target_folder,str(image_label),origin_name))

```

```python
#—————————————— valset recovery

dataset_path = 'ImageNet-1k-256'
N = 128
gray_flag = 1
save_as = 'png'
split = 'val'

if gray_flag == 1:
    split_real = split + '_gray_' + str(N)
else:
    split_real = split + '_' + str(N)

target_folder = os.path.join(dataset_path,split_real)

#find all val-***.parquet file
parquet_files = [os.path.join(root, file)
             for root, dirs, files in os.walk(dataset_path)
             for file in files if file.endswith('.parquet') and file.startswith(split)]

for file in parquet_files:
    parquet_data = pd.read_parquet(file)
    #print(parquet_data)
    for i,row in parquet_data.iterrows():
        single_pic_recover(row,target_folder,N,gray_flag,save_as)

        
```

```python
#—————————————— valset recovery

dataset_path = 'ImageNet-1k-256'
N = 256
gray_flag = 1
save_as = 'png'
split = 'test'

if gray_flag == 1:
    split_real = split + '_gray_' + str(N)
else:
    split_real = split + '_' + str(N)

target_folder = os.path.join(dataset_path,split_real)

#find all test-***.parquet file
parquet_files = [os.path.join(root, file)
             for root, dirs, files in os.walk(dataset_path)
             for file in files if file.endswith('.parquet') and file.startswith(split)]

for file in parquet_files[1]:
    parquet_data = pd.read_parquet(file)
    #print(parquet_data)
    for i,row in parquet_data.iterrows():
        single_pic_recover(row,target_folder,N,gray_flag,save_as)
```

```python
# Delete all files in the specified folder
import os

def delete_files_in_folder(folder):
    # Iterate over the given folder
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        # Check if it is a file
        if os.path.isfile(file_path) or os.path.islink(file_path):
            # Delete the file
            os.remove(file_path)
            # print(f'Deleted {file_path}')
        else:
            print(f'Skipped {file_path}')

# Replace with your directory path
folder_path = os.path.join('ImageNet-1k-256', 'test', '-1')
delete_files_in_folder(folder_path)
```
